```python
model = PPO.load("PPO_test3_modflow_multiInflow_coast")
env = gym.make('pumpingAllocation-v4')

#Pick scenerio
scenerio = 8

p = [0]*10
p[scenerio-1] = 1
#p = [0, 0.3, 0.3, 0.4,0]

env.river_p = p

SW, GW, Sup, D = [], [], [], []
deficit=[]
S, h, inflow = [], [], []
rewards = []
env.ifRandtt = False

for i in range(1):
    print('----------test',i+1,'start-------------')
    state = env.reset()
    total_step=0
    while True:
        action, _states = model.predict(state,deterministic=True )
        #print(action, _states)
        next_state, reward, done, info = env.step(action)
        #print(next_state,reward,done)
        
        #
        month = (total_step//3)%12+1
        #month = (state[3]+1)/2*11+1
        storage = (state[0]+1)/2*23500
        Zmin_before = (state[1]+1)/2*100-100
        Zmin_after = (next_state[1]+1)/2*100-100
        demand = (state[2]+1)/2*(3006.85-1201.95+400)+1300-200
        inf = (state[4]+1)/2*(30721.8-336.2+200)+336.2-200
        
        sw = action[0]*300
        gw = action[1]*216*240*20/10000
        
        print(f'{month:2.0f}, {total_step+1:2.0f}, :{storage:5.0f}, ():{Zmin_before:5.1f}{Zmin_after:4.1f}, Action:{action}, SWGW:[{sw:3.0f}, {gw:3.0f}], :{demand:4.0f}, Reward: {reward:.1f}')
        
        SW.append(sw)
        GW.append(gw)
        Sup.append(sw+gw)
        D.append(demand)
        deficit.append(max(0,demand-sw-gw))
        S.append(storage)
        h.append(Zmin_before)
        inflow.append(inf)
        rewards.append(reward)
        
        state = next_state
        total_step+=1

        if done:
            break
```

```python
import matplotlib.pyplot as plt
import pandas as pd
end=180

df = pd.DataFrame({'SW':SW[:end],
                   'GW':GW[:end],
                   'Supply':Sup[:end],
                   'Demand':D[:end],
                   'ith period of tendays':[i for i in range(1,end+1)],
                   'deficit':deficit[:end],
                   'S':S[:end],
                   'h':h[:end],
                   'In':inflow[:end],
                   'deficit_rate':[(b-a)/b*100 for a,b in zip(Sup[:end],D[:end])],
                   'reward':rewards[:end]
                  }) 


ax = df[['SW','GW']].plot(kind='bar', stacked=True, figsize=(18,6), width=0.7,
                          xlim=[1,180], ylim=[0,4100], xticks=[9*i for i in range(1,21)], fontsize=14)
df['Demand'].plot(kind='line', color='#FF0000', lw=2, xlim=[1,180],legend=True)
df['Supply'].plot(kind='line', style='g.-', lw=2, xlim=[1,180],legend=True)
ax.set_title('SW+GW_Scenerio_{}'.format(scenerio), fontsize=16)
ax.set_xlabel("tendays", fontsize=14)
ax.set_ylabel("m^3", fontsize=14)
ax.legend()
```
